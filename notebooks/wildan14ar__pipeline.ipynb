# Dicoding Submission MLOPS

Link Dataset: https://www.kaggle.com/datasets/gargmanas/sentimental-analysis-for-tweets

### Import library yang diperlukan

```python
import os
import tfx
from tfx.orchestration.experimental.interactive.interactive_context import InteractiveContext
from tfx.components import CsvExampleGen, StatisticsGen, SchemaGen, ExampleValidator
from tfx.components import Transform, Trainer, Evaluator, Pusher
from tfx.dsl.components.common.resolver import Resolver
from tfx.dsl.experimental import latest_blessed_model_resolver
from tfx.proto import trainer_pb2, pusher_pb2
from tfx.types import Channel
from tfx.types.standard_artifacts import Model, ModelBlessing
```

### Tentukan path direktori

```python
_pipeline_root = os.path.join(os.getcwd(), '../wildan14ar-pipeline')  # Folder untuk menyimpan artifact
_data_root = os.path.join(os.getcwd(), '../data')  # Folder data
_module_file_transform = os.path.join(os.getcwd(), '../modules/transform.py')
_module_file_trainer = os.path.join(os.getcwd(), '../modules/trainer.py')
_serving_model_dir = os.path.join(_pipeline_root, 'serving_model')
```

### Inisialisasi InteractiveContext

```python
context = InteractiveContext(pipeline_root=_pipeline_root)
```

### ExampleGen (membaca CSV)

```python
data_path = os.path.join(_data_root, 'data.csv')
example_gen = CsvExampleGen(input_base=_data_root)
context.run(example_gen)
```

### StatisticsGen

```python
statistics_gen = StatisticsGen(
    examples=example_gen.outputs['examples']
)
context.run(statistics_gen)
```

### SchemaGen

```python
schema_gen = SchemaGen(
    statistics=statistics_gen.outputs['statistics']
)
context.run(schema_gen)
```

### ExampleValidator

```python
example_validator = ExampleValidator(
    statistics=statistics_gen.outputs['statistics'],
    schema=schema_gen.outputs['schema']
)
context.run(example_validator)
```

### Transform

```python
transform = Transform(
    examples=example_gen.outputs['examples'],
    schema=schema_gen.outputs['schema'],
    module_file=_module_file_transform
)
context.run(transform)
```

### Trainer

```python
trainer = Trainer(
    module_file=_module_file_trainer,
    examples=transform.outputs['transformed_examples'],
    schema=schema_gen.outputs['schema'],
    transform_graph=transform.outputs['transform_graph'],
    train_args=trainer_pb2.TrainArgs(num_steps=2000),
    eval_args=trainer_pb2.EvalArgs(num_steps=500)
)
context.run(trainer)
```

### Resolver (untuk mengambil model terakhir yang sudah di-bless)

```python
model_resolver = Resolver(
    strategy_class=latest_blessed_model_resolver.LatestBlessedModelResolver,
    model=Channel(type=Model),
    model_blessing=Channel(type=ModelBlessing)  
).with_id('latest_blessed_model_resolver')
context.run(model_resolver)
```

### Evaluator

```python
from tensorflow_model_analysis.proto import config_pb2 as tfma_config

eval_config = tfma_config.EvalConfig(
    model_specs=[
        tfma_config.ModelSpec(label_key="label")
    ],
    metrics_specs=[
        tfma_config.MetricsSpec(
            metrics=[
                tfma_config.MetricConfig(
                    class_name="BinaryAccuracy",
                    threshold=tfma_config.MetricThreshold(
                        # Lebih singkat: langsung pakai dict
                        value_threshold={'lower_bound': {'value': 0.7}}
                    )
                )
            ]
        )
    ],
    slicing_specs=[tfma_config.SlicingSpec()]
)


evaluator = Evaluator(
    examples=example_gen.outputs['examples'],
    model=trainer.outputs['model'],
    eval_config=eval_config
)
context.run(evaluator)
```

### Pusher

```python
pusher = Pusher(
    model=trainer.outputs['model'],
    model_blessing=evaluator.outputs['blessing'],
    push_destination=tfx.proto.pusher_pb2.PushDestination(
        filesystem=pusher_pb2.PushDestination.Filesystem(
            base_directory=_serving_model_dir
        )
    )
)
context.run(pusher)
```
